```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import normalize
import glob
%matplotlib inline
```

```python
def create_dataframe(path):
    feature_names = []
    for unit_label in ["T", "RA", "LA", "RL", "LL"]:
        for sensor  in ["acc","gyro","mag"]:
            for position in ['X','Y','Z']:
                feature_names.append(unit_label + "_" + position + sensor)

    return pd.read_csv(path, names = feature_names)
```

```python
sit_df = create_dataframe('../data/a01/p1/s06.txt')
walk_df = create_dataframe('../data/a09/p2/s07.txt')
jump_df = create_dataframe('../data/a18/p3/s08.txt')
run_df = create_dataframe('../data/a12/p4/s09.txt')
cycle_df = create_dataframe('../data/a15/p5/s10.txt')
```
